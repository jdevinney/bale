```python
import json
import matplotlib.pyplot as plt
```

```python
# load the json file here. This is the file produced by running an app with the -j option
with open('gupc.json') as js:
    D = json.load(js)
```

```python
def get_plot_points(d):
    xlist = [int(i) for i in d]
    ylist = [float(d[i]) for i in d]
    return(xlist, ylist)

#fig,ax = plt.subplots()
agi_times = {}
exstack_times = {}
exstack2_times = {}
conveyor_times = {}
for record in D:
    if 'AGP' in record:
        agi_times[record['pes']]=record['AGP']
    if 'Exstack' in record:
        exstack_times[record['pes']]=record['Exstack']
    if 'Exstack2' in record:
        exstack2_times[record['pes']]=record['Exstack2']
    if 'Conveyor' in record:
        conveyor_times[record['pes']]=record['Conveyor']
if len(agi_times):
    x,y = get_plot_points(agi_times)
    plt.plot(x,y, marker='o', label='agi')
if len(exstack_times):
    x,y = get_plot_points(exstack_times)
    plt.plot(x,y, marker='o', label='exstack')
if len(exstack2_times):
    x,y = get_plot_points(exstack2_times)
    plt.plot(x,y, marker='o', label='exstack2')
if len(conveyor_times):
    x,y = get_plot_points(conveyor_times)
    plt.plot(x,y, marker='o',label='conveyor')
plt.ylabel('time (s)')
plt.xlabel('PEs')
plt.legend()
```
